### Churn by Contract Type

```python
df["Contract"].value_counts()
```

### Churn by Contract Type Analysis

```python
df.groupby("Contract")["Churn"].value_counts()
```

### Contract Type Insight

Customers with month-to-month contracts show significantly higher churn compared to customers with one-year and two-year contracts.

This suggests that long-term contracts improve customer retention and reduce churn risk.

```python
pd.crosstab(
    df["Contract"],
    df["Churn"],
    normalize="index"
) * 100
```

### Contract Type Business Insight

The churn rate for month-to-month customers is 42.71%, which is significantly higher than one-year (11.28%) and two-year (2.85%) contract customers.

This indicates that customers without long-term commitments are much more likely to leave the company. Encouraging customers to move to longer contract plans may help reduce churn.

### Churn by Customer Tenure

```python
df["tenure"].describe()
```

```python
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,72],
    labels=[
        "0-12 Months",
        "13-24 Months",
        "25-48 Months",
        "49-72 Months"
    ]
)
```

```python
pd.crosstab(
    df["tenure_group"],
    df["Churn"],
    normalize="index"
) * 100
```

### Customer Tenure Business Insight

Customers who have been with the company for less than 12 months show the highest churn rate (47.68%).

The churn rate consistently decreases as customer tenure increases, reaching only 9.51% for customers who have stayed between 49 and 72 months.

This suggests that customer retention efforts should focus heavily on the first year of the customer lifecycle.
